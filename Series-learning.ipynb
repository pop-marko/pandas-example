## Nums - Replacing NaN with numbers

```python
data = pd.read_csv('loan_risk_prediction_dataset.csv', usecols=['CreditScore']).squeeze('columns') ## pulling only CreditScore column for this example
data
```

```python
copied_data = data.copy() # Explicitly copying data, not giving pandas room for slip-ups
```

```python
mean_of_nums = copied_data.mean().round(2)

def replace_nan(num): # Func that goes into .apply() 
    if pd.isna(num):
        return mean_of_nums
    else:
        return num
        
copied_data.apply(replace_nan) 
```

## String Job - Filling strings instead of NaNs

```python
educations = pd.read_csv('loan_risk_prediction_dataset.csv', usecols=['Education']).squeeze('columns')
educations
```

```python
copy_edu = educations.copy()
```

```python
def fill_na_blanks(education):
    if pd.isna(education):
        return 'not specified'.title()
    else:
        return education.title()

copy_edu.apply(fill_na_blanks).value_counts()
```

## Ratings labeled with numbers 1-3 for risk

```python
netflix = pd.read_csv('NetFlix.csv', usecols=['rating']).squeeze('columns')
netflix.value_counts()
```

```python
def ratings_risk(rating):
    high_risk = ['TV-MA','R','NR','UR']
    if rating in high_risk:
        return 3
    elif rating == 'TV-14' or rating == 'PG-13':
        return 2
    else:
        return 1
rating_numbers = netflix.apply(ratings_risk).value_counts()
rating_numbers
```
